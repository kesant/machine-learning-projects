## Flujo de Creación de un Modelo de Clasificación

### 1. **Definición del Problema:**
   - Clarifica el problema que estás tratando de resolver y determina si es un problema de clasificación.
   - Define las clases que deseas predecir.

### 2. **Preparación del Entorno y Carga de Datos:**
   - Importación de bibliotecas necesarias (por ejemplo, scikit-learn, pandas, numpy).
   - Carga del conjunto de datos.


### 3. **Exploración de Datos:**

  - Exploración y comprensión de la distribución de datos.
  - Visualización de datos para obtener información relevante.
  - Tratamiento de valores atípicos y valores faltantes si es necesario.

### 4. **Preprocesamiento de Datos:**

- Declaración de vectores de características y variable objetivo:

    - Vectores de Características: Seleccionar las variables independientes (características) que se utilizarán para entrenar el modelo.
    - Variable Objetivo: Identificar la variable dependiente que el modelo intentará predecir.

- División de datos:
    - Divide el conjunto de datos en tres partes: entrenamiento, validación y prueba.
    - El conjunto de entrenamiento se utiliza para entrenar el modelo.
    - El conjunto de validación se utiliza para ajustar los hiperparámetros del modelo.
    - El conjunto de prueba se utiliza al final para evaluar el rendimiento del modelo.

- Tratamiento de Desbalanceo de Datos:

  - Si el conjunto de datos presenta un desbalance en las clases (por ejemplo, una clase tiene muchos más ejemplos que otra), se deben aplicar técnicas de tratamiento de desbalanceo para mejorar la capacidad predictiva del modelo.
  - Comparación de Métodos para Abordar el Desbalance en los datasets

| Técnica                  | Ventajas                                                   | Desventajas                                             | Casos de Uso                                              |
|--------------------------|------------------------------------------------------------|---------------------------------------------------------|-----------------------------------------------------------|
| **Sobre-muestreo**       | - Aumenta la cantidad de ejemplos de la clase minoritaria.  | - Posible sobreajuste si se aplica de manera excesiva.   | - Cuando hay datos insuficientes en la clase minoritaria. |
|                          | - Mejora la capacidad del modelo para aprender patrones de la clase minoritaria. |                                                         | - Cuando la clase minoritaria es pequeña y representativa. |
|                          | - Puede ser efectivo cuando la clase minoritaria es pequeña y representativa. |                                                         |                                                           |
| **Muestreo Estratificado** | - Mantiene una proporción equilibrada de clases en cada estrato. | - Puede ser computacionalmente costoso en conjuntos de datos grandes. | - Cuando hay múltiples clases y se quiere mantener representatividad. |
|                           | - Evita que ciertas clases queden subrepresentadas en la muestra. |                                                         | - Útil cuando hay clases con pocas instancias en comparación con otras. |
|                           | - Útil cuando hay clases con pocas instancias en comparación con otras. |                                                         |                                                           |
| **Sub-muestreo**          | - Reduce el riesgo de sobreajuste en la clase mayoritaria.    | - Pérdida de información al eliminar instancias.        | - Cuando hay datos en exceso en la clase mayoritaria.    |
|                           | - Puede mejorar el rendimiento en casos de desbalance extremo. | - Puede no funcionar bien si se eliminan instancias clave. |                                                           |
|                           | - Reduce el costo computacional al trabajar con menos datos. |                                                         |                                                           |
| **Ponderación de Clases** | - Permite asignar pesos diferentes a las clases durante el entrenamiento. | - Requiere ajuste adecuado de los pesos para obtener buenos resultados. | - Cuando se quiere dar más importancia a la clase minoritaria sin modificar el tamaño del conjunto de datos. |
|                           | - Ayuda al modelo a enfocarse más en la clase minoritaria durante el entrenamiento. |                                                         |                                                           |


- Feature Engineering:

  - Manipular las características para mejorar la calidad de la información que proporcionan al modelo de machine learning.

- Feature Scaling:

  - Ajustar las escalas de las características para asegurar que todas contribuyan de manera equitativa al modelo.


### 5. **Selección del Modelo:**


   - Elige un modelo de clasificación basado en la naturaleza del problema y los datos disponibles (por ejemplo, regresión logística, SVM, árboles de decisión, Random Forest, redes neuronales, etc.).

### 7.Evaluación del Modelo con Datos de Prueba:
  - Utiliza el conjunto de prueba (que el modelo no ha visto antes) para evaluar el rendimiento final del modelo.

#### Metricas de evaluacion
1. Accuracy (Precisión):

    * Definición: Mide la proporción de predicciones correctas realizadas por el modelo.
    * Interpretación: Un valor alto de accuracy indica que el modelo realiza un buen trabajo en general. Sin embargo, no siempre refleja la performance en cada clase, especialmente en casos con clases desbalanceadas.
    * Limitaciones: No toma en cuenta la confusión entre clases, lo que puede ser un problema si el modelo tiene un bajo rendimiento en una clase específica.

2. Recall (Sensibilidad o Remembranza):

    * Definición: Mide la proporción de casos positivos reales que fueron correctamente identificados por el modelo.
    * Interpretación: Un valor alto de recall indica que el modelo es efectivo para identificar la mayoría de los casos positivos.
    * Utilidad: Es útil cuando es crucial detectar todos los casos positivos, como en un caso de diagnóstico médico.

3. F1-Score:

    * Definición: Es una métrica ponderada que combina la precisión y el recall en una sola medida.
    * Interpretación: Un valor alto de F1-score indica un buen equilibrio entre precisión y recall, lo que es ideal para la mayoría de las aplicaciones.
    * Utilidad: Es una métrica útil para comparar diferentes modelos, especialmente cuando se tiene un conjunto de datos desbalanceado.

4. AUC (Área Bajo la Curva ROC):

    * Definición: Mide la capacidad del modelo para distinguir entre casos positivos y negativos.
    * Interpretación: Un AUC de 0.5 indica un rendimiento aleatorio, mientras que un AUC de 1 indica un rendimiento perfecto.
    * Utilidad: Es útil para comparar diferentes modelos y para evaluar la capacidad del modelo para generalizar a nuevos datos.
    
5. Matriz de confusión:

    * Definición: La matriz de confusión es una tabla que permite visualizar el desempeño de un modelo de clasificación. Muestra la cantidad de casos reales para cada clase en comparación con las predicciones del modelo.
    * Interpretación:Una matriz de confusión ideal tendría la mayoría de los valores en la diagonal principal (TP(verdaderos positivos) y TN(verdaderos negativos)).Un alto número de FP(falsos positivos) y FN(falsos negativos) indica que el modelo está confundiendo las clases.
    * Utilidad:La matriz de confusión es una herramienta útil para evaluar el rendimiento de un modelo de clasificación y para identificar áreas de mejora.Se puede utilizar para comparar diferentes modelos y para seleccionar el modelo más adecuado para una aplicación específica.

##### Suposición


Supongo que los datos faltan de forma completamente aleatoria (MCAR). Hay dos métodos que se pueden utilizar para imputar valores faltantes. Uno es la imputación de la media o mediana y el otro es la imputación de una muestra aleatoria. Cuando hay valores atípicos en el conjunto de datos, debemos utilizar la imputación de la mediana. Por lo tanto, utilizaré la imputación de la mediana porque la imputación de la mediana es robusta ante los valores atípicos.


Imputaré los valores faltantes con las medidas estadísticas apropiadas de los datos, en este caso la mediana.

```python
def max_value(df3, variable, top):
    return np.where(df3[variable]>top, top, df3[variable])

for df3 in [X_train, X_test]:
    df3['Rainfall'] = max_value(df3, 'Rainfall', 3.2)
    df3['Evaporation'] = max_value(df3, 'Evaporation', 21.8)
    df3['WindSpeed9am'] = max_value(df3, 'WindSpeed9am', 55)
    df3['WindSpeed3pm'] = max_value(df3, 'WindSpeed3pm', 57)
```

```python
# One-Hot Encoding para Location en ambos conjuntos
encoder_location = OneHotEncoder(sparse=False, drop='first')

# Concatenar los conjuntos antes de aplicar la transformación
X_combined = pd.concat([X_train[['Location']], X_test[['Location']]], axis=0)

# Aplicar One-Hot Encoding
encoded_location_combined = encoder_location.fit_transform(X_combined[['Location']])

# Dividir nuevamente los conjuntos después de la transformación
encoded_location_train = encoded_location_combined[:len(X_train)]
encoded_location_test = encoded_location_combined[len(X_train):]

# Asignar las columnas codificadas de nuevo a los conjuntos originales
X_train = pd.concat([X_train, pd.DataFrame(encoded_location_train, columns=encoder_location.get_feature_names_out(['Location']))], axis=1)
X_test = pd.concat([X_test, pd.DataFrame(encoded_location_test, columns=encoder_location.get_feature_names_out(['Location']))], axis=1)

# Label Encoding para WindGustDir, WindDir9am, WindDir3pm en ambos conjuntos
label_encoder = LabelEncoder()
for col in ['WindGustDir', 'WindDir9am', 'WindDir3pm']:
    X_train[col] = label_encoder.fit_transform(X_train[col])
    X_test[col] = label_encoder.transform(X_test[col])

# One-Hot Encoding para RainToday en ambos conjuntos
encoder_rain_today = OneHotEncoder(sparse=False, drop='first')
encoded_rain_today_combined = encoder_rain_today.fit_transform(pd.concat([X_train[['RainToday']], X_test[['RainToday']]], axis=0))

X_train = pd.concat([X_train, pd.DataFrame(encoded_rain_today_combined[:len(X_train)], columns=encoder_rain_today.get_feature_names_out(['RainToday']))], axis=1)
X_test = pd.concat([X_test, pd.DataFrame(encoded_rain_today_combined[len(X_train):], columns=encoder_rain_today.get_feature_names_out(['RainToday']))], axis=1)
```

#### Matriz de confusión <a class="anchor" id="15"></a>
Una matriz de confusión es una herramienta para resumir el desempeño de un algoritmo de clasificación. Una matriz de confusión nos dará una imagen clara del rendimiento del modelo de clasificación y los tipos de errores producidos por el modelo. Nos da un resumen de predicciones correctas e incorrectas desglosadas por cada categoría. El resumen se presenta en forma de tabla.


Son posibles cuatro tipos de resultados al evaluar el desempeño de un modelo de clasificación. Estos cuatro resultados se describen a continuación: -


**Verdaderos positivos (TP)**: los verdaderos positivos ocurren cuando predecimos que una observación pertenece a una determinada clase y la observación en realidad pertenece a esa clase.


**Verdaderos Negativos (TN)** – Los Verdaderos Negativos ocurren cuando predecimos que una observación no pertenece a una determinada clase y la observación en realidad no pertenece a esa clase.


**Falsos positivos (FP)**: los falsos positivos ocurren cuando predecimos que una observación pertenece a una determinada clase, pero la observación en realidad no pertenece a esa clase. Este tipo de error se llama **Error tipo I.**



**Falsos negativos (FN)**: los falsos negativos ocurren cuando predecimos que una observación no pertenece a una determinada clase, pero la observación en realidad pertenece a esa clase. Este es un error muy grave y se llama **Error de tipo II.**



```python
# Imprime la Matriz de Confusión y córtala en cuatro pedazos

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_test)

print('Matriz de confusión\n\n', cm)

print('\nVerdaderos positivos(TP) = ', cm[0,0])

print('\nVerdaderos Negativos(TN) = ', cm[1,1])

print('\nFalsos positivos(FP) = ', cm[0,1])

print('\nFalsos negativos(FN) = ', cm[1,0])
```

```python
# visualizar matriz de confusión con mapa de calor marino

cm_matrix = pd.DataFrame(data=cm, columns=['Positivo real:1', 'Negativo real:0'],
                                  index=['Predecir positivo:1', 'Predecir negativo:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

#### ROC-AUC


**ROC AUC** significa **Característica operativa del receptor: área bajo la curva**. Es una técnica para comparar el rendimiento del clasificador. En esta técnica, medimos el "área bajo la curva (AUC)". Un clasificador perfecto tendrá un ROC AUC igual a 1, mientras que un clasificador puramente aleatorio tendrá un ROC AUC igual a 0,5.

Entonces, **ROC AUC** es el porcentaje del gráfico ROC que está debajo de la curva.

```python
# compute ROC AUC

from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_pred_test)

print('ROC AUC : {:.4f}'.format(ROC_AUC))
```

##### Comentarios


- ROC AUC es un resumen numérico único del rendimiento del clasificador. Cuanto mayor sea el valor, mejor será el clasificador.

- ROC AUC de nuestro modelo se acerca a 1. Por lo tanto, podemos concluir que nuestro clasificador hace un buen trabajo al predecir si lloverá mañana o no.

##### Suposición


Supongo que los datos faltan de forma completamente aleatoria (MCAR). Hay dos métodos que se pueden utilizar para imputar valores faltantes. Uno es la imputación de la media o mediana y el otro es la imputación de una muestra aleatoria. Cuando hay valores atípicos en el conjunto de datos, debemos utilizar la imputación de la mediana. Por lo tanto, utilizaré la imputación de la mediana porque la imputación de la mediana es robusta ante los valores atípicos.


Imputaré los valores faltantes con las medidas estadísticas apropiadas de los datos, en este caso la mediana.

```python
# imputar valores faltantes

for col in df.columns:
      col_median=df[col].median()
      df[col].fillna(col_median, inplace=True)
```
