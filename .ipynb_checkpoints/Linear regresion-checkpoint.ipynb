```python
#usamos one harding coding
df[["no","yes"]]=pd.get_dummies(df["smoker"])
```

```python
df.head()
```

```python
x=df.loc[:,["age","bmi","no","yes"]]
y=df["charges"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
model= LinearRegression()#inicializar el modelo
```

```python
model.fit(X_train,y_train)#encuentra los coeficientes para encontrar la salida y
```
